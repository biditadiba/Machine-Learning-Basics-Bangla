# মেশিন লার্নিং অ্যালগরিদম: নেইভ বেইজ ক্লাসিফায়ার(Naive Bayes Classifier)

নেইভ বেইজ ক্লাসিফায়ার একটি ক্লাসিফিকেশন টাইপ মেশিন লার্নিং অ্যালগরিদম। এই অ্যালগরিদমটি মূলত বেইজ থিওরাম এর উপর ভিত্তি করে তৈরি হয়েছে। <br>


# বেইজ থিওরাম

বেইজ থিওরাম হচ্ছে জটিল জটিল ইভেন্টের কন্ডিশনাল প্রবাবিলিটি নির্ণয়য়ের জন্য একটি অসাধারন গানিতিক তত্ত্ব। প্রবাবিলিটির দুনিয়ায় বেইজ থিওরাম  গুরুত্ব ব্যাপক। আমরা যখন মেশিন লার্নিং শিখবো সেখানেও দেখবো বিভিন্ন ধরনের ক্লাসিফিকেশন প্রব্লেম সল্ভ করার জন্য বেইজ থিওরাম এর প্রয়োগ রয়েছে। নেইভ বেইজ অ্যালগোরিদম নামে মেশিন লার্নিং-এ আলাদা একটি অ্যালগোরিদমই রয়েছে বেইজ থিওরাম এর উপর ভিত্তি করে। <br>
বেইজ থিওরাম এর মাধ্যমে মূলত হিডেন ইভেন্টের প্রবাবিলিটি বের করা যায়, যেটা আমাদের আগে থেকে জানা থাকে না।  ধরে নেই A এবং B দুইটি ইভেন্ট। B ইভেন্ট ঘটলে A ইভেন্ট ঘটার প্রবাবিলিটি কত সেটাই বেইজ বেইজ থিওরাম এর মাধ্যমে নির্ণয় করা হয়। <br>
 <br>
 P(A) = A ইভেন্ট ঘটার প্রবাবিলিটি<br>
 P(B) = B ইভেন্ট ঘটার প্রবাবিলিটি<br>
 P(B|A) = A ইভেন্ট ঘটলে B ইভেন্ট ঘটার প্রবাবিলিটি <br>
 P(A|B)=  B ইভেন্ট ঘটলে A ইভেন্ট ঘটার প্রবাবিলিটি <br>
 <br>
উদাহরন - ধরে নিন জানুয়ারি মাসের কোন এক শুক্রবার আপনি পরিবার এবং বন্ধুদের নিয়ে বাইরে কোথাও পিকনিকে যাবার প্ল্যান করেছেন। শুক্রবার দিন সকালে ঘুম থেকে উঠেই দেখলেন আকাশটা মেঘলা, আর আপনার কপালেও দুশ্চিন্তার ছাপ , বৃষ্টি হলে আপনার সব প্ল্যান বাতিল ! এই অবস্থায় আপনার কাছে পূর্বের কিছু ইনফরমেশন আছে, এই ইনফরমেশন গুলোকে কাজে লাগিয়ে আপনি জানতে চাচ্ছেন  আজকে বৃষ্টি হবার প্রবাবিলিটি আসলে কত ? <br>
পূর্বের ইনফরমেশন:<br>
৪০% দিনেই সকাল বেলায় আকাশ মেঘলা থাকে<br>
৫০% ক্ষেত্রে মেঘলা আকাশ থাকলেই বৃষ্টিপাত হয়<br>
১০% জানুয়ারি মাসে সাধারনত ৩ দিনের বেশী বৃষ্টি হয় না <br>
সুতরাং আমরা লিখতে পারি,<br>
P(Rain) = 10%<br>
P(Cloud|Rain)  = 50%<br>
P(Cloud) = 40%<br>
<br>
সুতরাং আপনি জেনে গেলেন আজ বৃষ্টি হবার প্রবাবিলিটি মাত্র ১২.৫% , তাই মেঘ দেখে ভয় না পেলেও চলবে, পিকনিক করা যেতেই পারে কারন বৃষ্টি হবার সম্ভাবনা একেবারেই কম ! <br>

# নেইভ বেইজ ক্লাসিফায়ার

সহজ কথায় বেইজ থিওরাম হচ্ছে একটি ঘটনা (X) ঘটলে আরেকটি ঘটনা (Y)  ঘটার প্রবাবিলিটি বের করার পদ্ধতি । যেমন আকাশে মেঘ দেখা গেলে বৃষ্টি হবার সম্ভাবনা। বেইজ থিওরামকে গানিতিক ভাবে এভাবে লেখা যায়,<br>
<br>
উপরের সমীকরণটি একটি সরল বেইজ সমীকরন যা কেবল মাত্র একটি কন্ডিশনাল ইভেন্টের ক্ষেত্রে প্রবাবিলিটি নির্ণয়ে ব্যাবহার হয়। বাস্তব ক্ষেত্রে বেশিরভাগ ডেটাসেটই মাল্টিভেরিয়েট হয়ে থাকে , সেক্ষেত্রে সমীকরণটাও একটু জটিল হয়ে যায়। তখন আমরা সমীকরণটি এভাবে লিখতে পারি, <br>
<br>
নেইভ বেইজ ক্লাসিফায়ার এর কিছু বৈশিষ্ট্য :<br>
এটি খুব সহজেই ইমপ্লিমেন্ট করা যায় এবং তুলনামূলক ভাবে দ্রুত কাজ করে।  <br>
ছোট ডেটাসেটের ক্ষেত্রেও ভালো কাজ করে।  <br>
অন্যান্য অ্যালগরিদমের তুলনায় অ্যাকুরেসি কিছুটা কম দেয়।  <br>
নেইভ বেইজে সকল এট্রিবিউটসকে মিউচুয়ালি ইন্ডিপেনডেন্ট ধরা হয় কিন্তু বাস্তব জগতে অনেক ঘটনার ক্ষেত্রেই এমনটি হয় না। <br>

# অ্যাকুরেসি নির্ণয় 

সকল ধরনের ক্লাসিফিকেশন অ্যালগরিদম এর  অ্যাকুরেসি নির্ণয় করা হয় কনফিউশন ম্যাট্রিক্স এর মাধ্যমে। কনফিউশন ম্যাট্রিক্স চার ধরনের ভ্যালুর সমন্বয়ে গঠিত হয়, এগুলো হোল,<br>
ট্র্রু পজিটিভ- মডেল পজিটিভ হিসাবে প্রিডিক্ট করেছে এবং তা সঠিক, যেমন প্রেগন্যান্ট মহিলাকে প্রেগন্যান্ট হিসাবে প্রেডিক্ট করা। <br>
ফলস পজিটিভ- মডেল পজিটিভ হিসাবে প্রিডিক্ট করেছে কিন্তু সেটা ভুল, যেমন একজন পুরুষকে প্রেগন্যান্ট হিসাবে প্রিডিক্ট করা। <br>
ট্র্রু নেগেটিভ- মডেল নেগেটিভ হিসাবে প্রেডিক্ট করেছে এবং সেটা সঠিক , যেমন একজন পুরুষকে প্রেগন্যান্ট নয় হিসাবে প্রিডিক্ট করা। <br>
ফলস নেগেটিভ - মডেল নেগেটিভ হিসাবে প্রিডিক্ট করেছে কিন্তু সেটা ভুল, যেমন একজন প্রেগন্যান্ট মহিলাকে , প্রেগন্যান্ট নয় এমন প্রিডিক্ট করা। <br>
 <br>
কনফিউশন ম্যাট্রিক্স এর False Positive কে বলা হয় Type 1 Error এবং False Negative কে বলা হয় Type 2 Error । কনফিউশন ম্যাট্রিক্স এর মাধ্যমে মডেলের অ্যাকুরেসি নির্ণয়ের জন্য কিছু পদ্ধতি অবলম্বন করা হয়। TP, TN, FP, FN এগুলো থেকেই মূলত মডেলের পারফরম্যান্স নির্ণয় করা হয়। <br>
<br>
Accuracy ((TP+TN)/Total)-  ওভারঅল মডেল কতখানি সঠিক ভাবে প্রিডিক্ট করতে পারে। <br>
Sensitivity/Recall (TP/(TP + FN))- পজিটিভকে পজিটিভ হিসাবে প্রিডিক্ট করতে পারার অনুপাত। <br>
Precision ( TP/(TP+FP))-  মডেলর সঠিক পজিটিভ অনুমান এবং মোট (সঠিক এবং ভুল) পজিটিভ অনুমানের অনুপাত। <br>
Specificity (TN/(TN + FP))- নেগেটিভকে নেগেটিভ হিসাবে  প্রিডিক্ট করতে পারার অনুপাত। <br>
<br>
নেইভ বেইজ ক্লাসিফায়ার বেশ কিছু ধরনের হতে পারে। আমরা যে ধরনের প্রবলেম সল্ভ করবো তার উপর ভিত্তি করেই ক্লাসিফায়ার নির্ধারণ করবো। <br>
Multinational Naive Bayes: প্রেডিকশন ক্লাস একাধিক হলে ব্যাবহার করা হয়। যেমন ডকুমেন্ট ক্লাসিফিকেশন। <br>
Bernoulli Naive Bayes:   বুলিয়ান ভ্যারিয়েবল এর ক্ষেত্রে ব্যাবহার করা হয়। <br>
Gaussian Naive Bayes:  কন্টিনিউয়াস ভেরিয়েবল ফিচার হিসাবে থাকলে ব্যাবহার করা হয় ।<br>
