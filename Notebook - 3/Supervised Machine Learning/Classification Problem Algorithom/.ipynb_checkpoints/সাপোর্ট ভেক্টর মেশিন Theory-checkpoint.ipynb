# মেশিন লার্নিং অ্যালগরিদম: সাপোর্ট ভেক্টর মেশিন(Support Vector Machine)

সাপোর্ট ভেক্টর মেশিন অ্যালগরিদমটি ক্লাসিফিকেশন এবং রিগ্রেশন উভয় ধরনের সমস্যার ক্ষেত্রেই ব্যবহার করা যায়। তবে অপেক্ষাকৃত জটিল ধরনের ছোট বা মাঝারি ক্লাসিফিকেশন ডেটাসেটের জন্য এসভিএম বেশ জনপ্রিয়। এই অ্যালগরিদমে ডেটা পয়েন্টকে একটি হাইপারপ্লেইন এর মাধ্যমে আলাদা করা হয়, এক্ষেত্রে হাইপারপ্লেইনটি কেমন হবে সেটা কার্নেল এর মাধ্যমে নির্ধারণ করা হয়। <br>
<br>
সাপোর্ট ভেক্টর মেশিন এর বিভিন্ন উপাদান:<br>
সাপোর্ট ভেক্টরস- ডেটাসেটের বিভিন্ন ক্লাসের প্রান্ত দেশে থাকা ডেটা পয়েন্ট গুলো যা ক্লাসগুলোর মধ্যে পার্থক্য গড়ে দেয়, এমন ডেটাই হচ্ছে সাপোর্ট ভেক্টরস।<br>
হাইপারপ্লেইন- হাইপারপ্লেইন হচ্ছে একটি সিদ্ধান্ত মূলক রেখা বা লাইন যা একাধিক শ্রেনির ডেটার ভেতরে পার্থক্য নির্ধারণ করে। <br>
মার্জিন- মার্জিন হচ্ছে একাধিক শ্রেনির ডেটার ভেতরের দূরত্ব । মার্জিন নির্ধারণ করা হয় সাপোর্ট ভেক্টরের মধ্যকার দূরত্বের মাধ্যমে। দুইটি ক্লাসের ভেতরের মার্জিন যত বেশী হবে তাকে তত বেশী ভালো মার্জিন বলা হবে। <br>
সাপোর্ট ভেক্টর মেশিন এর বিভিন্ন ধরনের কার্নেল: <br><br>
সাধারন স্কাটার প্লটে আমরা যদি একাধিক ভ্যারিয়েবলকে প্লট করি অনেক ক্ষেত্রেই সেই প্লটের মাধ্যমে দুই বা ততধিক ডেটা ক্লাসকে আলাদা করা যায় না। সাপোর্ট ভেক্টর মেশিন এর  কার্নেল হচ্ছে একটি বিশেষ ধরনের কৌশল যা লোয়ার ডাইমেনশনাল ডেটা কে হাইয়ার ডাইমেনশনাল স্পেসে কনভার্ট করতে পারে এবং এর ফলে ক্লাসগুলোর মধ্যকার পার্থক্য স্পস্ট হয়। <br>
লিনিয়ার কার্নেল- লিনিয়ার কার্নেল হচ্ছে একটি লিনিয়ার লাইন বা সরলরেখা যার মাধ্যমে ডেটাপয়েন্টকে আলাদা করা হয়। <br>
পলিনমিয়াল কার্নেল - পলিনমিয়াল লাইন হচ্ছে নন-লিনিয়ার লাইন বা বক্ররেখা। অনেক ক্ষেত্রেই লিনিয়ার লাইনের মাধ্যমে ডেটাপয়েন্ট আলাদা করা যায় না এসব ক্ষেত্রে পলিনমিয়াল কার্নেল ব্যবহার করতে হয়। <br>
রেডিয়াল বেসিস ফাংশন কার্নেল-  রেডিয়াল বেসিস ফাংশনও একধরনের নন-লিনিয়ার ফাংশন। এই ফাংশনটিই সাপোর্ট ভেক্টর মেশিন এর সবথেকে জনপ্রিয় ফাংশন। এই কার্নেল কোন ইনপুটকে ইনফাইনাইট ডাইমেনশনাল স্পেসে  ম্যাপ করতে পারে। <br><br>
লিনিয়ার প্রবলেমের ক্ষেতে সাপোর্ট ভেক্টর মেশিন যেভাবে কাজ করে:<br>
প্রথমে একাধিক হাইপারপ্লেইন জেনারেট করা হয় , নিচের প্রথম ছবিতে ( বাম পাশে ) আমরা দেখতে পাচ্ছি ক্লাসিফিকেশনের জন্য তিনটি হাইপারপ্লেইন জেনারেট করা হয়েছে। এরপর প্রতিটি হাইপারপ্লেইনের ক্লাসিফিকেশন এরর ক্যালকুলেট করা হয়, দেখা হয় কোন হাইপারপ্লেইনের জন্য এরর সবথেকে কম এবং সেই হাইপারপ্লেইনটিকেই ক্লাসিফিকেশনের জন্য বাছাই করা হয়। <br>
<br>
নন-লিনিয়ার প্রবলেমের ক্ষেতে সাপোর্ট ভেক্টর মেশিন যেভাবে কাজ করে:<br> 
অনেক ডেটাসেটকেই ক্লাসিফিকেশনের জন্য লিনিয়ারলি সেপারেট করার মত অবস্থায় পাওয়া যাবে না , সেক্ষেত্রে নন-লিনিয়ার কার্নেল ডেটাকে হাইয়ার ডাইমেশনে ট্রান্সফরমেশন করে নেয় ফলে এটি ক্লাসিফিকেশনের উপযোগী হয়ে উঠে। <br>
<br>
সাপোর্ট ভেক্টর মেশিন এর কিছু বৈশিষ্ট্য: <br>
হাই ডাইমেনশনাল স্পেসের জন্য সাপোর্ট ভেক্টর মেশিন অত্যন্ত কার্যকরী। <br>
এই অ্যালগরিদমটি তুলনামূলক কম মেমোরি ব্যবহার করে। <br>
অনেক বড় ডেটাসেটের জন্য সাপোর্ট ভেক্টর মেশিন তেমন কার্যকরী নয়। <br>
ডেটাসেটে ওভারল্যাপিং ক্লাস এবং নয়েজ থাকলে মডেল সহজেই প্রভাবিত হয়।<br>
<br>
সাপোর্ট ভেক্টর রিগ্রেশনের মধ্যে ব্যবহৃত দুটি মুখ্য হাইপারপ্যারামিটার হলো "গামা" (gamma) এবং "সি" (C).<br>
গামা (gamma) হলো একটি কর্ণেল ফাংশনের প্যারামিটার, যা মডেলের কম্প্লেক্সিটি নির্ধারণ করে। গামা বেশ বড় হলে মডেল কম্প্লেক্স হবে, যা অন্যদিকে সমস্যার অতিরিক্ত অনুমান বা ওভারফিটিং (overfitting) এর জন্য উপযুক্ত হতে পারে। সমস্যার সম্ভাব্য উপর নির্ভর করে, আপনার ডেটা সেটের সংখ্যকতা এবং পর্যাপ্ত মানুষের জ্ঞানের উপর।<br>
সি (C) হলো রেগুলারাইজেশনের প্যারামিটার যা রিগ্রেশন মডেলের ক্ষমতা নির্ধারণ করে। সি মান যত বেশি হবে, মডেলটি ক্ষমতাশালী হবে, যা অন্যদিকে ওভারফিটিং এর জন্য উপযুক্ত হতে পারে। সমস্যার ধরন ও ডেটা সেটের সম্প্রদায়ের উপর নির্ভর করে, এবং আপনার পছন্দ অনুযায়ী ব্যবহার করা যেতে পারে।<br>
