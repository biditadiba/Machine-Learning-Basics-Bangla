# মেশিন লার্নিং অ্যালগরিদম: ডিসিশন ট্রি (Decision Tree)

<br>
উপরে একটা স্পিড লিমিট সাইন এবং একটা ডায়াগ্রাম দেওয়া। ডায়াগ্রামের সবার উপরের নোডে দেওয়া “গাড়ির স্পিড >= ৫০”। এখন আমরা জিজ্ঞেস করি “গাড়ির স্পিড কি ৫০ এর সমান বা বেশি?” যদি উত্তর হয় “হ্যাঁ”, তবে গাড়ির স্পিড কমায় আনতে হবে। যদি উত্তর হয় “না”, তবে গাড়ির স্পিড যেমন আছে, ওই স্পিডে চললেই হবে। <br>
এইখানে একটা হ্যাঁ-না সূচক প্রশ্ন করে আমরা নির্ধারণ করছি গাড়ির স্পিড কি কমাবো নাকি চলতে থাকবে একই স্পিডে। এইটা একটা সিম্পল ডিসিশন ট্রি-এর উদাহরণ।<br>

Decision Tree: ডিসিশন ট্রি একটা সুপার্ভাইসড মেশিন লার্নিং অ্যালগরিদম যা ক্লাসিফিকেশন এবং রিগ্রেশন দুই ক্ষেত্রেই ব্যবহৃত হয়। তবে ক্লাসিফিকেশনে বেশি ব্যবহৃত হয়। এইটা একটা ট্রি-স্ট্রাকচার্ড ক্লাসিফায়ার। অর্থাৎ এর একটা রুট থাকে, সেইখান থেকে ব্রাঞ্চ তৈরি হয়, সেইখান থেকে লিফ নোড বের হয়।<br>

ডিসিশন ট্রি অ্যালগরিদম কিভাবে কাজ করে? <br>
কোন প্রদত্ত ডাটাসেটের ক্লাস প্রেডিক্ট করার জন্য অ্যালগরিদমটি ডিসিশন ট্রি-এর রুট নোড থেকে কাজ শুরু করে। রুট নোডের অ্যাট্রিবিউটের সাথে তুলনা করে (“গাড়ির স্পিড কি ৫০ এর চেয়ে বেশি?”) ডিসিশন নেওয়া হয় যে কোন ব্রাঞ্চের দিকে আগাবে (যদি উত্তর হয় “হ্যাঁ”, তবে গাড়ির স্পিড কমায় আনতে হবে। যদি উত্তর হয় “না”, তবে গাড়ির স্পিড যেমন আছে, ওই স্পিডে চললেই হবে)। এইভাবে পরবর্তী যে নোড সিলেক্ট হয়, সেই নোডের অ্যাট্রিবিউটের সাথে তুলনা করে আবার ঠিক করে এখন কোন ব্রাঞ্চের দিকে যাবে। এইভাবে প্রত্যেক ধাপে সে নোডের অ্যাট্রিবিউটের সাথে তুলনা করে ঠিক করতে থাকে এখন সে কোন ব্রাঞ্চের দিকে যাবে যতক্ষণ পর্যন্ত না লিফ নোডে পৌঁছায়। লিফ নোডে পৌঁছান মানে আমরা যেইটা ক্লাসিফাই করতে চাচ্ছিলাম, সেইটা ক্লাসিফাই করা হয়ে গিয়ছে। পুরো প্রসেসটা নিচে দেওয়া হল: <br>

স্টেপ১: রুট নোড থেকে শুরু করতে হবে যেইটায় এখন পুরো ডাটাসেট রয়েছে। ধরি এই রুট নোড ’S’। <br>
স্টেপ২: প্রত্যেক ইটারেশনে S-এ অব্যবহৃত অ্যাট্রিবিউটগুলোর অ্যাট্রিবিউট সিলেকশন মেজার (ASM) ইনডেক্সগুলো ক্যালকুলেট করে। <br>
স্টেপ৩: ক্যালকুলেশন থেকে বেস্ট অ্যাট্রিবিউট সিলেক্ট করে। <br>
স্টেপ৪: সিলেক্ট করা বেস্ট অ্যাট্রিবিউটের ভিত্তিতে S-কে স্প্লিট করে সাবসেট অফ ডাটা বানায়। <br>
স্টেপ৫: রিকার্সিভলি প্রত্যেক সাবসেটে এই অ্যালগরিদম চলতে থাকে, প্রত্যেকবার অব্যবহৃত অ্যাট্রিবিউটগুলো কন্সিডার করে। <br>


# Construction of Decision Tree: Gain ratio

<br>
এইখানে আমাদের অ্যাট্রিবিউট ও তাদের ভ্যালু:<br>
Age: Youth, Middle-aged, Senior<br>
Income: High, Medium, Low<br>
Student: Yes, No<br>
Credit rating: Fair, Excellent<br>

আমাদের টার্গেট ক্লাস – Buys Computer: No, Yes<br>

এখন টার্গেট ক্লাসের এন্ট্রপি বের করি-<br>
 <br>
E(Buys Computer) = – Pyes log2(Pyes) – Pno log2(Pno)<br>

                 = – 9/14 log2 (9/14) – 5/14 log2 (5/14) <br>
                 
                 = 0.94<br>

এখন আমরা অ্যাট্রিবিউটগুলোর গেইন রেশিও বের করব। প্রথমে Age এর গেইন রেশিও বের করি। <br>
<br>
প্রত্যেক ব্রাঞ্চের জন্য এন্ট্রপি বের করি – <br>

E(Youth) = – 2/5 log2(2/5) – 3/5 log2(3/5) = 0.97<br>
E(Middle-aged) = -4/4 log2(4/4) – 0/4 log2(0/4) = 0<br>
E(Senior) = – 3/5 log2(3/5) – 2/5 log2(2/5) = 0.97<br>

এখন টার্গেটের সাপেক্ষে Outlook এর এন্ট্রপি বের করি – <br>

E(Buys Computer, Age) = P(Youth)E(Youth) + P(Middle-aged)E(Middle-aged) + P(Senior)E(Senior) <br>

                      = 5/14 * 0.97 + 4/14 * 0 + 5/14 * 0.97 = 0.69<br>

Age এর গেইন রেশিও –<br>
Information Gain (Age) = E(Buys Computer) – E(Buys Computer, Age)   <br>                                                                                
                       = 0.94 – 0.69 = 0.25<br>

SplitInfo(Age) = -5/14 log2 (5/14) – 4/14 log2(4/14) – 5/14 log2 (5/14) 

                = 1.58<br>

GainRatio(Age) = Information Gain(Age) / SplitInfo (Age) 

               = 0.25 / 1.58 <br>
               
               = 0.16<br>

এইভাবে প্রত্যেক অ্যাট্রিবিউটের গেইন রেশিও বের করলে পাই:<br>

GainRatio(Age) = 0.16<br>
GainRatio(Income) = 0.02<br>
GainRatio(Student) = 0.15<br>
GainRatio(Credit rating) = 0.05<br>

তাহলে আমরা দেখতে পাচ্ছি যে ‘Age’ অ্যাট্রিবিউটের গেইন রেশিও সবচেয়ে বেশি, তাই এইটাই হবে আমাদের রুট নোড। আমরা আরও দেখেছি যে Middle-aged ব্রাঞ্চের এন্ট্রপি 0, সুতরাং এইটাই হবে আমাদের লিফ নোড। বাকি দুই ব্রাঞ্চের এন্ট্রপি 0 এর থেকে বেশি, তাই আরও স্প্লিটিং হবে।<br>

<br>

এখন আমরা Youth ব্রাঞ্চের স্প্লিটিং করব। আমাদের এখন অ্যাট্রিবিউট বাকি আছে – Income, Student, Credit rating। আমরা তাইলে এখন Youth এর সাপেক্ষে Income এর ইনফরমেশন গেইন বের করি।<br>

<b>
    
প্রত্যেক ব্রাঞ্চের জন্য এন্ট্রপি বের করি –<b>

E(Low) = -1/1 log2 (1/1) – 0/1 log2 (0/1) = 0<b>
    
E(Medium) = -1/2 log2(1/2) – 1/2 log2(1/2) = 1<b>
    
E(High) = -0/2 log2(0/2) – 2/2 log2(2/2) = 0<b>
    
এখন Youth এর ব্রাঞ্চের জন্য Income এর এন্ট্রপি –<b>
    
E(Youth, Income) = P(Low)E(Low) + P(Medium)E(Medium) + P(High)E(High)  <b> 
    
                 = 1/5 * 0 + 2/5 * 1 + 2/5 * 0 = 0.4 <b>

Income এর গেইন রেশিও – <b>
    
Information Gain (Income) = E(Youth) – E(Youth, Income) 
                          
                          = 0.97 – 0.4 = 0.57 <b>
    
SplitInfo (Income) = -1/5 log2(1/5) – 2/5 log2 (2/5) – 2/5 log2(2/5)
                   
                   = 1.52 <b>
    
GainRatio (Income) = Information Gain (Income) / SplitInfo (Income) 
    
                   = 0.57 / 1.52 = 0.38 <b>

অনুরূপভাবে,<b>
    
GainRatio(Student) = 1 <b>
    
GainRatio(Credit rating) = 0.23 <b>

তাহলে দেখা যাচ্ছে Student এর গেইন রেশিও সবচেয়ে বেশি। সুতরাং এইটা হবে আমাদের পরবর্তী নোড। ব্রাঞ্চগুলোর এন্ট্রপি বের করলে দেখব যে এইগুলাই আমাদের রুট নোড হবে। <b>
<br>
    
অনুরূপভাবে আমরা পরবর্তী ডিসিশন নোড পাই হল Credit rating। এর ব্রাঞ্চের দুইটার এন্ট্রপি 0, তাই আর স্প্লিটিং হবেনা। আমরা এইখানেও লিফ নোড পেয়ে গেলাম। আমাদের ডিসিশন ট্রি গঠন সম্পূর্ণ হল।<br>
