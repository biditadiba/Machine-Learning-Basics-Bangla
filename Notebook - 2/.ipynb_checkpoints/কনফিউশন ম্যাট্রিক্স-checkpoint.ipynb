# কনফিউশন ম্যাট্রিক্স

সকল ধরনের ক্লাসিফিকেশন অ্যালগরিদম এর  অ্যাকুরেসি নির্ণয় করা হয় কনফিউশন ম্যাট্রিক্স এর মাধ্যমে। কনফিউশন ম্যাট্রিক্স চার ধরনের ভ্যালুর সমন্বয়ে গঠিত হয়, এগুলো হোল,<br>
ট্র্রু পজিটিভ- মডেল পজিটিভ হিসাবে প্রিডিক্ট করেছে এবং তা সঠিক, যেমন প্রেগন্যান্ট মহিলাকে প্রেগন্যান্ট হিসাবে প্রেডিক্ট করা। <br>
ফলস পজিটিভ- মডেল পজিটিভ হিসাবে প্রিডিক্ট করেছে কিন্তু সেটা ভুল, যেমন একজন পুরুষকে প্রেগন্যান্ট হিসাবে প্রিডিক্ট করা। <br>
ট্র্রু নেগেটিভ- মডেল নেগেটিভ হিসাবে প্রেডিক্ট করেছে এবং সেটা সঠিক , যেমন একজন পুরুষকে প্রেগন্যান্ট নয় হিসাবে প্রিডিক্ট করা। <br>
ফলস নেগেটিভ - মডেল নেগেটিভ হিসাবে প্রিডিক্ট করেছে কিন্তু সেটা ভুল, যেমন একজন প্রেগন্যান্ট মহিলাকে , প্রেগন্যান্ট নয় এমন প্রিডিক্ট করা। <br>
<br>
কনফিউশন ম্যাট্রিক্স এর False Positive কে বলা হয় Type 1 Error এবং False Negative কে বলা হয় Type 2 Error । কনফিউশন ম্যাট্রিক্স এর মাধ্যমে মডেলের অ্যাকুরেসি নির্ণয়ের জন্য কিছু পদ্ধতি অবলম্বন করা হয়। TP, TN, FP, FN এগুলো থেকেই মূলত মডেলের পারফরম্যান্স নির্ণয় করা হয়।<br>
<br>
Accuracy ((TP+TN)/Total)-  ওভারঅল মডেল কতখানি সঠিক ভাবে প্রিডিক্ট করতে পারে। <br>
Sensitivity/Recall (TP/(TP + FN))- পজিটিভকে পজিটিভ হিসাবে প্রিডিক্ট করতে পারার অনুপাত। <br>
Precision ( TP/(TP+FP))-  মডেলর সঠিক পজিটিভ অনুমান এবং মোট (সঠিক এবং ভুল) পজিটিভ অনুমানের অনুপাত। <br>
Specificity (TN/(TN + FP))- নেগেটিভকে নেগেটিভ হিসাবে  প্রিডিক্ট করতে পারার অনুপাত। <br>
<br>

# F1 Score:

F1 score is a machine learning evaluation metric that measures a model's accuracy. It combines the precision and recall scores of a model. The accuracy metric computes how many times a model made a correct prediction across the entire dataset. It ranges from 0 to 1, with a higher value indicating better performance. An F1 score of 1 indicates perfect precision and recall, while an F1 score of 0 indicates that the model is not able to correctly predict any positive instances. <br>
The F1 score is calculated as follows:<br>

F1 score = 2 * (precision * recall) / (precision + recall)<br>

**F1 score ** ** Interpretation **<br>

0.9 Very good<br>
0.8 - 0.9 Good<br>
0.5 - 0.8 OK<br>
< 0.5 Not good<br>

# Receiver Operating Characteristic:

ROC (Receiver Operating Characteristic) কার্ভ হল একটি গ্রাফ যা মডেলের কিভাবে বিভিন্ন থ্রেশহোল্ড মান ব্যবহার করে সঠিকভাবে পজিটিভ এবং নেগেটিভ ইনস্ট্যান্স শনাক্ত করছে তা প্রদর্শিত করে। এই কার্ভের প্রতিটি বিন্দুতে থ্রেশহোল্ড মান পরিবর্তিত করলে পজিটিভ এবং নেগেটিভ ক্লাসিফিকেশনের জন্য মডেলের সেন্সিটিভিটি এবং স্পেসিফিসিটি পরিবর্তিত হবে।

# Area Under the Curve

AUC (Area Under the Curve) হল রক কার্ভের নিচে অংশের মাপ। এটি সত্ত্বাধিক্যের পরিমাপ হিসাবে ব্যবহৃত হয়। AUC এর মান 0 থেকে 1 এর মধ্যে হয় এবং এর মানের উচ্চতা মডেলের পারফরমেন্সকে প্রদর্শিত করে। যত বেশি AUC মান, ততটা মডেলের পারফরমেন্স উচ্চ হবে।

মেশিন লার্নিং এ ROC এবং AUC কার্ভ ব্যবহার করা হয় মডেল পারফরমেন্স মাপার জন্য, যেমন ক্লাসিফিকেশন টাস্কে, ডিসক্রিট প্রেডিকশনে এবং বিভিন্ন কাস্টম থ্রেশহোল্ড মান ব্যবহার করে।

```python
#Come here after finishing Notebook3 
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split as tts
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_curve
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('heart.xlsx')
df.head()
```

```python
df.DEATH_EVENT.value_counts()
```

```python
sns.countplot(df.DEATH_EVENT)
```

```python
sns.countplot(x='diabetes',hue='DEATH_EVENT',data=df)
```

```python
sns.countplot(x='high_blood_pressure',hue='DEATH_EVENT',data=df)
```

```python
df.corr()
```

```python
plt.figure(figsize=(18,12))
plt.title('Correlation for every column')
sns.heatmap(df.corr(),annot=True)
```

```python
x = df.drop('DEATH_EVENT',axis=1)
y=df[['DEATH_EVENT']]
x.head()
```

```python
y.head()
```

```python
xtrain,xtest,ytrain,ytest = tts(x,y,test_size=0.3,random_state=42)
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
```

```python
pred = dt.predict(xtest)#predicted
pred
```

```python
np.array(ytest)#real
```

```python
print(classification_report(ytest,pred))
```

```python
cm = confusion_matrix(ytest,pred)
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted Result')
plt.ylabel('Actual values')
plt.title('confusion matrix')
```

```python
#ROC
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

# Compute the predicted probabilities for the positive class
y_pred_prob = dt.predict_proba(xtest)[:, 1]

# Compute the false positive rate, true positive rate, and thresholds
fpr, tpr, thresholds = roc_curve(ytest, y_pred_prob)

# Plot the ROC curve
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```

```python
accuracy_score(ytest,pred)
```

```python
dt.score(xtest,ytest)
```
