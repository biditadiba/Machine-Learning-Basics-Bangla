# Encoding Techniques in Machine Learning


সাধারণ ভাষায় এনকোডিং হচ্ছে মেশিন লার্নিং এর এমন একটি পদ্ধতি যেখানে স্ট্রিং ভ্যালু গুলোকে ইন্টিজার  ভ্যালুতে পরিণত করা হয় <br>

১.লেবেল এনকোডিং: লেবেল এনকোডিং  ক্যাটেগরিক্যাল ডেটা প্রসেস করার একটি পদ্ধতি। এই পদ্ধতিতে আমারা ক্যাটেগরিক্যাল ভ্যারিয়েবেলের সকল ইউনিক ভ্যালুকে আলাদা আলাদা নাম্বার দিয়ে প্রকাশ করতে পারি। ধরুন আপনার কাছে একটা  ডেটাসেট আছে সেখানে মানুষের  জেন্ডার হিসাবে পুরুষ বা স্ত্রী লেখা আছে। আমরা লেবেল এনকোডিং এর মাধ্যমে পুরুসদেরকে ১ এবং স্ত্রীদেরকে ২ দ্বারা প্রকাশ করতে পারি।<br>
<br>
এখানে আমরা এরিয়া কলামে দেখতে পাই যে তিনটি ইউনিক ভ্যালু আছে সেগুলো হচ্ছে Dahka,Ctg এবং Rangpur। তো এই তিনটা জায়গার নামকে যদি আমরা অ্যাসেন্ডিং অর্ডারে সাজাই তাহলে সেটা আসে হচ্ছে Ctg, Dhaka, Rangpur। <br>
After Label encoding,<br>
Ctg = 0<br>
Dhaka = 1<br>
Rangpur = 2<br>
লেবেল এনকোডিং এর মাধ্যমে আমরা প্রতিটি ইউনিক নামকে আলাদা আলাদা নাম্বার দ্বারা প্রকাশ করেছি।

২.ওয়ান হট এনকোডিং: এটি একটি জনপ্রিয় এনকডার যার মাধ্যমে ক্যাটেগরিক্যাল ভ্যারিয়্যাবলকে আলাদা আলাদা কলামে কনভার্ট করা যায় এবং প্রতিটি কলামে বুলিয়ান  ট্রু / ফলস বা 0/1 দ্বারা তাদের উপস্থিতি প্রকাশ করা যায়।

#  One Hot Encoding

```python
df3 = df.copy()
df3.head()
```

```python
#making smoker column values into integer corresponding to yes and no. when yes in active tehn it is 1 and no is 0
dummies = pd.get_dummies(df3['smoker']).astype(int)
dummies.head(5) 
```

```python
df3 = pd.concat([df3,dummies],axis='columns')#adding column of dummies
df3.head() 
```

```python
df3.drop(['smoker'],axis='columns',inplace=True)
df3.head()
```

```python
#We can also do this using loop if there are more columns
cols = ['gender','region']
if df3 is not None:
    for i in cols:
        dummies = pd.get_dummies(df3[i]).astype(int)
        df3 = pd.concat([df3, dummies], axis='columns')
        df3.drop([i], axis='columns', inplace=True)

df3.head()
```

# Ordinal Encoding

```python
df4 = df.copy()
df4.head()
```
